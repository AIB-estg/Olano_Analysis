O Objetivo será fazer a previsão da procura, mais concretamente da quantidade diária de paletes transportada pela empresa.

A escolha entre desenvolver um modelo de previsão de paletes para cada cliente/localidade ou para a quantidade diária total depende dos objetivos da empresa e da complexidade dos dados.

Uma abordagem interessante seria combinar os dois modelos:

- Um modelo global para prever a procura total diária da empresa.
- Modelos individuais para clientes ou localidades chave, onde os padrões são mais distintos ou onde existe uma grande variabilidade na procura.


Sugestão:
**Análise de variabilidade:** Antes de decidir, seria interessante verificar se há diferenças significativas nos padrões de procura entre os clientes e localidades. Isso pode ser feito com uma análise de séries temporais de cada cliente/localidade.

Para analisar a variabilidade entre as regiões e identificar sazonalidades distintas, vamos seguir os seguintes passos:

**Agrupamento de Dados por Região:** Agrupamos os dados por região (localidade) e data (data_carga), somando o número de paletes.
**Análise de Sazonalidade:** Usamos métodos de decomposição de séries temporais para extrair as tendências sazonais e identificar padrões distintos entre as regiões.
**Visualização de Duas Regiões com Sazonalidades Distintas: **Escolhemos duas regiões com padrões sazonais diferentes para mostrar que seus comportamentos são distintos.

**Passo 1:** Analisar a Variabilidade por Região
O código abaixo agrupa as paletes por localidade e data, depois calcula a soma diária de paletes para cada região

```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Carregando o dataset - substitua o caminho do arquivo pelo seu
# data = pd.read_excel('/content/seu_arquivo.xlsx')

# Mostrando as primeiras linhas para verificar o formato
data.head()

# Limpando e preparando os dados
# Certifique-se de que a coluna data_carga é do tipo datetime
data['data_carga'] = pd.to_datetime(data['data_carga'])

# Agrupando os dados por localidade e data, somando as paletes por dia para cada localidade
region_data = data.groupby(['localidade', 'data_carga'])['n_paletes'].sum().reset_index()

# Verificando as primeiras linhas do novo dataframe
region_data.head()

# Criando um pivô para facilitar a análise por localidade
region_pivot = region_data.pivot(index='data_carga', columns='localidade', values='n_paletes')

# Verificando as primeiras linhas do pivô
region_pivot.head()

```

**Conclusão:**
Com base nos padrões observados nas decomposições de séries temporais, um modelo por região parece ser a escolha mais apropriada. # Próximos passos - Comparar os dois tipos de modelo LSTM - Global e por região

Vamos usar o Keras Tuner para afinar os hiperparâmetros de cada modelo e comparar as métricas de ajuste e previsão (MSE, RMSE, MAE, e R²) em cada caso.

**Passos:**
1. *Pré-processamento:* Preparar os dados para o modelo LSTM (normalizar, separar conjunto de treino e teste).

2. *Modelo LSTM Global:* Criar um modelo LSTM que prevê a demanda global (para todas as regiões).

3. *Modelo LSTM por Região:* Criar modelos LSTM específicos para Lisboa e Abrantes.

*Afinação dos Hiperparâmetros:* Usar o Keras Tuner para afinar os hiperparâmetros de cada modelo.

4. *Avaliação:* Avaliar o desempenho de cada modelo com métricas de ajuste e previsão.


```python
#### PAsso 1 ######
# Importando as bibliotecas
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Carregando os dados
# data = pd.read_excel('/content/seu_arquivo.xlsx')

# Convertendo a data
data['data_carga'] = pd.to_datetime(data['data_carga'])

# Agrupando os dados por localidade e data, somando as paletes por dia
region_data = data.groupby(['localidade', 'data_carga'])['n_paletes'].sum().reset_index()

# Criando pivô para facilitar a análise por localidade
region_pivot = region_data.pivot(index='data_carga', columns='localidade', values='n_paletes').fillna(0)

# Normalizando os dados (usando MinMaxScaler)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(region_pivot)

# Criando uma função para transformar em séries temporais
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i + n_steps])
        y.append(data[i + n_steps])
    return np.array(X), np.array(y)

# Definindo o número de timesteps (dias de sequência)
n_steps = 30
X, y = create_sequences(scaled_data, n_steps)

# Separando em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
#### PASSO 2 ####

from kerastuner.tuners import RandomSearch
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Função para criar o modelo LSTM global
def build_global_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32),
                   return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(hp.Float('dropout_rate', 0.2, 0.5, step=0.1)))
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32)))
    model.add(Dense(y_train.shape[1]))

    model.compile(optimizer=Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 5e-4])),
                  loss='mse')
    return model

# Usando Keras Tuner para afinar hiperparâmetros
tuner_global = RandomSearch(
    build_global_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='lstm_global_tuning')

# Treinando o modelo global com Keras Tuner
tuner_global.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test))

# Selecionando o melhor modelo
best_global_model = tuner_global.get_best_models(num_models=1)[0]
```

```python
#### PASSO 3 ######

# Filtrando dados para Lisboa e Abrantes
lisboa_data = region_pivot[['LISBOA']].dropna()
abrantes_data = region_pivot[['ABRANTES']].dropna()

# Normalizando e criando sequências para cada cidade
lisboa_scaled = scaler.fit_transform(lisboa_data)
abrantes_scaled = scaler.fit_transform(abrantes_data)

X_lisboa, y_lisboa = create_sequences(lisboa_scaled, n_steps)
X_abrantes, y_abrantes = create_sequences(abrantes_scaled, n_steps)

# Separando treino e teste para Lisboa e Abrantes
X_train_lisboa, X_test_lisboa, y_train_lisboa, y_test_lisboa = train_test_split(X_lisboa, y_lisboa, test_size=0.2, shuffle=False)
X_train_abrantes, X_test_abrantes, y_train_abrantes, y_test_abrantes = train_test_split(X_abrantes, y_abrantes, test_size=0.2, shuffle=False)

# Função para criar o modelo LSTM para uma cidade
def build_region_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32),
                   return_sequences=True, input_shape=(X_train_lisboa.shape[1], X_train_lisboa.shape[2])))
    model.add(Dropout(hp.Float('dropout_rate', 0.2, 0.5, step=0.1)))
    model.add(LSTM(units=hp.Int('units', min_value=32, max_value=128, step=32)))
    model.add(Dense(1))

    model.compile(optimizer=Adam(hp.Choice('learning_rate', [1e-2, 1e-3, 5e-4])),
                  loss='mse')
    return model

# Afinar hiperparâmetros para Lisboa
tuner_lisboa = RandomSearch(
    build_region_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='lstm_lisboa_tuning')

tuner_lisboa.search(X_train_lisboa, y_train_lisboa, epochs=50, validation_data=(X_test_lisboa, y_test_lisboa))

# Melhor modelo para Lisboa
best_lisboa_model = tuner_lisboa.get_best_models(num_models=1)[0]

# Afinar hiperparâmetros para Abrantes
tuner_abrantes = RandomSearch(
    build_region_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='lstm_abrantes_tuning')

tuner_abrantes.search(X_train_abrantes, y_train_abrantes, epochs=50, validation_data=(X_test_abrantes, y_test_abrantes))

# Melhor modelo para Abrantes
best_abrantes_model = tuner_abrantes.get_best_models(num_models=1)[0]
```

```python
#### PASSO 4 #######

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Função para calcular métricas
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"MSE: {mse}, RMSE: {rmse}, MAE: {mae}, R2: {r2}")
    return mse, rmse, mae, r2

# Avaliar o modelo global
print("Métricas do Modelo Global:")
evaluate_model(best_global_model, X_test, y_test)

# Avaliar o modelo de Lisboa
print("Métricas do Modelo Lisboa:")
evaluate_model(best_lisboa_model, X_test_lisboa, y_test_lisboa)

# Avaliar o modelo de Abrantes
print("Métricas do Modelo Abrantes:")
evaluate_model(best_abrantes_model, X_test_abrantes, y_test_abrantes)
```

Para comparar previsões de curto, médio e longo prazo entre os três modelos (global, Lisboa e Abrantes), vamos também incluir intervalos de confiança para medir a incerteza dessas previsões.

Os passos serão os seguintes:

1- Curto, Médio e Longo Prazo: Gerar previsões para 10, 30 e 90 dias para os modelos Global, Lisboa e Abrantes.

3- Comparação: Comparar as previsões entre os três modelos para diferentes períodos, incluindo os intervalos de confiança.

**Passo 1:** Cálculo dos Intervalos de Confiança
Os intervalos de confiança serão baseados nos resíduos das previsões no conjunto de teste. Usaremos os resíduos para calcular a incerteza associada às previsões.

Função para Gerar Previsões com Intervalo de Confiança
Aqui está o código para prever com intervalos de confiança. Ele pode ser reutilizado para o modelo global, Lisboa e Abrantes.

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def forecast_with_confidence_interval(model, X_test, n_days=10, confidence_level=1.96):
    predictions = []
    current_input = X_test[-1]  # Última observação real como ponto de partida

    for i in range(n_days):
        # Previsão para o próximo dia
        next_pred = model.predict(current_input.reshape(1, X_test.shape[1], X_test.shape[2]))[0]
        predictions.append(next_pred)

        # Atualizar input para incluir a previsão
        current_input = np.append(current_input[1:], [next_pred], axis=0)

    # Calcular os resíduos do conjunto de teste
    residuals = model.predict(X_test) - y_test[:len(X_test)]

    # Calcular erro padrão dos resíduos
    mse = mean_squared_error(y_test[:len(X_test)], model.predict(X_test))
    standard_error = np.sqrt(mse)

    # Intervalo de confiança
    lower_bound = np.array(predictions) - confidence_level * standard_error
    upper_bound = np.array(predictions) + confidence_level * standard_error

    # Plotando as previsões com intervalos de confiança
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, n_days + 1), predictions, label='Previsão', color='blue')
    plt.fill_between(range(1, n_days + 1), lower_bound, upper_bound, color='gray', alpha=0.3, label='Intervalo de Confiança')
    plt.title(f'Previsão de {n_days} dias com Intervalo de Confiança ({confidence_level}σ)')
    plt.xlabel('Dias')
    plt.ylabel('Número de Paletes')
    plt.legend()
    plt.show()

    return predictions, lower_bound, upper_bound
```

**Passo 2:** Gerar Previsões com Intervalo de Confiança para Global, Lisboa e Abrantes
Usaremos essa função para os três modelos: Global, Lisboa e Abrantes, e para os três períodos (curto, médio e longo prazo). Vamos agora gerar as previsões e compará-las.

```python
# Previsão de Curto Prazo (10 dias) com Intervalo de Confiança para o Modelo Global
global_short_predictions, global_lower_bound_short, global_upper_bound_short = forecast_with_confidence_interval(
    best_global_model, X_test, n_days=10, confidence_level=1.96)

# Previsão de Curto Prazo para o Modelo de Lisboa
lisboa_short_predictions, lisboa_lower_bound_short, lisboa_upper_bound_short = forecast_with_confidence_interval(
    best_lisboa_model, X_test_lisboa, n_days=10, confidence_level=1.96)

# Previsão de Curto Prazo para o Modelo de Abrantes
abrantes_short_predictions, abrantes_lower_bound_short, abrantes_upper_bound_short = forecast_with_confidence_interval(
    best_abrantes_model, X_test_abrantes, n_days=10, confidence_level=1.96)
```

Passo 3: Comparação das Previsões para Diferentes Períodos
Agora que geramos as previsões com intervalos de confiança para todos os três modelos (Global, Lisboa, Abrantes), podemos comparar as previsões de curto, médio e longo prazo entre eles.

Comparando as Previsões para Curto, Médio e Longo Prazo

```python
def compare_forecasts_with_confidence(global_preds, global_lb, global_ub, lisboa_preds, lisboa_lb, lisboa_ub,
                                      abrantes_preds, abrantes_lb, abrantes_ub, n_days=10):
    plt.figure(figsize=(12, 6))

    # Previsão Global
    plt.plot(range(1, n_days + 1), global_preds, label='Global', color='blue')
    plt.fill_between(range(1, n_days + 1), global_lb, global_ub, color='blue', alpha=0.2)

    # Previsão Lisboa
    plt.plot(range(1, n_days + 1), lisboa_preds, label='Lisboa', color='green')
    plt.fill_between(range(1, n_days + 1), lisboa_lb, lisboa_ub, color='green', alpha=0.2)

    # Previsão Abrantes
    plt.plot(range(1, n_days + 1), abrantes_preds, label='Abrantes', color='red')
    plt.fill_between(range(1, n_days + 1), abrantes_lb, abrantes_ub, color='red', alpha=0.2)

    plt.title(f'Comparação de Previsões para {n_days} Dias com Intervalos de Confiança')
    plt.xlabel('Dias')
    plt.ylabel('Número de Paletes')
    plt.legend()
    plt.show()

# Comparação para Curto Prazo (10 dias)
compare_forecasts_with_confidence(global_short_predictions, global_lower_bound_short, global_upper_bound_short,
                                  lisboa_short_predictions, lisboa_lower_bound_short, lisboa_upper_bound_short,
                                  abrantes_short_predictions, abrantes_lower_bound_short, abrantes_upper_bound_short,
                                  n_days=10)

# Comparação para Médio Prazo (30 dias)
compare_forecasts_with_confidence(global_medium_predictions, global_lower_bound_medium, global_upper_bound_medium,
                                  lisboa_medium_predictions, lisboa_lower_bound_medium, lisboa_upper_bound_medium,
                                  abrantes_medium_predictions, abrantes_lower_bound_medium, abrantes_upper_bound_medium,
                                  n_days=30)

# Comparação para Longo Prazo (90 dias)
compare_forecasts_with_confidence(global_long_predictions, global_lower_bound_long, global_upper_bound_long,
                                  lisboa_long_predictions, lisboa_lower_bound_long, lisboa_upper_bound_long,
                                  abrantes_long_predictions, abrantes_lower_bound_long, abrantes_upper_bound_long,
                                  n_days=90)
```
